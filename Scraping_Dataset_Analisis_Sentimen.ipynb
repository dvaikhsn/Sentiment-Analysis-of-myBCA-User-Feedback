### **Import Library yang dibutuhkan**

```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

### **Aplikasi yang mau di scarp**

```python
from google_play_scraper import Sort, reviews_all

scrap_apk = reviews_all(
    'com.bca.mybca.omni.android',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT
)
```

###**Menyimpan hasil scarping**

```python
import csv

with open('ulasan my-bca.csv', mode='w', newline='', encoding='utf-8') as file:
  writer = csv.writer(file)
  writer.writerow(['Reviews'])
  for review in scrap_apk:
    writer.writerow([review['content']])
```

### **Loading Dataset**

```python
apk_review_df = pd.DataFrame(scrap_apk)
apk_review_df.to_csv('ulasan my-bca.csv', index=False)
```
